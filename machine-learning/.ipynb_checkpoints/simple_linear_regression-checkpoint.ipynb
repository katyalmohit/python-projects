```python
#Display data using pandas
df = pd.read_csv("./FuelConsumption.csv")
df.head()
```

```python
#To summarise the data
df.describe()
```

```python
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head(10)
```

```python
viz = cdf[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB', 'CO2EMISSIONS']]
#To plot histogram
viz.hist(color='brown')
plt.show()
```

```python
#To plot graph
plt.scatter(cdf.FUELCONSUMPTION_COMB, cdf.CO2EMISSIONS, color='green')
plt.xlabel('FUELCONSUMPTION_COMB')
plt.ylabel('CO2 Emissions')
plt.show()
```

```python
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS, color='blue')
plt.xlabel('Engine Size')
plt.ylabel('CO2 Emissions')
plt.show()
```

```python
plt.scatter(cdf.CYLINDERS, cdf.CO2EMISSIONS, color='green')
plt.xlabel('Cylinders')
plt.ylabel('CO2 Emissions')
plt.show()
```

```python
# Training and testing dataset
# Here, we have considered only 80% of the value for training
# Remaining 20% values are left for testing
msk = np.random.rand(len(df))<0.8 
#Upper code ensures that the rows are selected at random and have 80% of total values
train = cdf[msk] #It will consider 80% of values
test = cdf[~msk] #It will consider the remaining values i.e. 20%
```

```python
#Plotting selected data for training
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='blue')
plt.xlabel('Engine Size')
plt.ylabel('CO2 Emissions')
plt.show()
```

```python
#Training data

#Import necessary libraries
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])

#To find the coefficient and intercept for best fit curve
regr.fit(train_x, train_y)

#Printing coefficient and intercept
print("Coefficients: ",regr.coef_)
print("Intercept: ",regr.intercept_)
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], 'red')
#first parameter is x
#second parameter is y
# y = mx + c
#Here, m is regr.coef_[0][0] and c is regr.intercept_[0]
plt.xlabel('Engine Size')
plt.ylabel('CO2 Emissions')
plt.show()
```

```python
# Error evalution
# Three different model evaluation:
# 1. Mean Absolute Error, 2. Mean Squared Error, 3. Root Mean Squared Error

from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])
#prediction
test_y_ = regr.predict(test_x)

#Displaying error
print("Mean Absolute Error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Mean Squared Error: %.2f" % np.mean((test_y_ - test_y)**2))
print("R2 Score: %.2f" %r2_score(test_y_, test_y))
#The higher the R-squared value, the better the model fits your data. 
#The best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse).
```

```python
# Doing it by taking another column
train_x = train[["FUELCONSUMPTION_COMB"]]
train_y = train[["CO2EMISSIONS"]]

regr.fit(train_x, train_y)

print("Coefficients: ",regr.coef_)
print("Intercept: ",regr.intercept_)
```

```python
# Plotting the best fit curve
plt.scatter(train.FUELCONSUMPTION_COMB, train.CO2EMISSIONS)
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], 'r')
plt.xlabel("FUELCONSUMPTION_COMB")
plt.ylabel("CO2 Emissions")
plt.show()
```

```python
predictions = regr.predict(train_x)
# Error evaluation
print("Mean Absolute Error: %.2f" %np.mean(np.absolute(predictions - train_y)))
print("Mean Squared Error: %.2f" %np.mean((predictions - train_y)**2))
print("R2 Score: %.2f" %r2_score(predictions, train_y))
```
